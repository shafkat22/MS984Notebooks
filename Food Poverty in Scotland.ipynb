```python
import pandas as pd
```

```python
xls = pd.ExcelFile('food_dataset2.xls')
df = pd.read_excel(xls, 'Scotland')
df
```

```python
df1 = df.iloc[[0,7,8,9,10,11,12]]
df1 = df1.drop(['Food Group'], axis=1)
df1 = df1.rename(columns={"Food Category": "Food Group"})
df1
```

```python
df = df.drop(['Food Category'], axis=1)
df = df.append(df1)
df = df.drop(['RSE indicator(a)', '% change since 201819', 'sig(b)', 'trend since 201819(c)'], axis=1)
df
```

```python
df = df.dropna()
df = df.drop([8, 10, 18, 19, 20, 22])
df
```

```python
df2 = df.loc[[0]]
df2[df2.select_dtypes(include=['number']).columns] *= 1.04
df2['Units'].replace('ml', 'g', inplace=True)
df2
```

```python
df3 = df.loc[[4]]
df3[df.select_dtypes(include=['number']).columns] *= 63
df3['Units'].replace('no.', 'g', inplace=True)
df3
```

```python
df = df.drop([0])
df = df.drop([4])
```

```python
df = df.append(df2)
df = df.append(df3)
```

```python
df['Food Group'].replace('Carcase meat', 'Fish & Meat', inplace=True)
df['Food Group'].replace('Non-carcase meat and meat products', 'Fish & Meat', inplace=True)
df['Food Group'].replace('Fish', 'Fish & Meat', inplace=True)
df['Food Group'].replace('Sugar and preserves', 'Added Sugars', inplace=True)
df['Food Group'].replace('Confectionery', 'Added Sugars', inplace=True)
df['Food Group'].replace('Cakes, buns and pastries', 'Carbs', inplace=True)
df['Food Group'].replace('Bread', 'Carbs', inplace=True)
df['Food Group'].replace('Flour', 'Carbs', inplace=True)
df['Food Group'].replace('Biscuits and crispbreads', 'Carbs', inplace=True)
df['Food Group'].replace('Other cereals and cereal products', 'Carbs', inplace=True)
df['Food Group'].replace('Fresh and processed potatoes', 'Carbs', inplace=True)
df['Food Group'].replace('Fresh and processed vegetables, excluding potatoes', 'Vegetables', inplace=True)
df['Food Group'].replace('Fresh and processed fruit', 'Fruit', inplace=True)
df['Food Group'].replace('Cheese', 'Eggs & Dairy', inplace=True)
df['Food Group'].replace('Eggs', 'Eggs & Dairy', inplace=True)
df['Food Group'].replace('Milk and milk products excluding cheese', 'Eggs & Dairy', inplace=True)
df
```

```python
df = df.groupby('Food Group',as_index=False).sum()
df
```

```python
# add subplots
import numpy as np
import matplotlib.pyplot as plt

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, sharex='col', sharey='row', figsize=(14, 10))

ax1.bar(df['Food Group'],df[201516], 0.2, label='2015/16')
ax2.bar(df['Food Group'],df[201617], 0.2, label='2016/17')
ax3.bar(df['Food Group'],df[201718], 0.2, label='2017/18')
ax4.bar(df['Food Group'],df[201819], 0.2, label='2018/19')
ax5.bar(df['Food Group'],df[201920], 0.2, label='2019/20')
ax6.bar(df['Food Group'],df[202021], 0.2, label='2020/21')

ax1.legend(loc="upper right")
ax2.legend(loc="upper right")
ax3.legend(loc="upper right")
ax4.legend(loc="upper right")
ax5.legend(loc="upper right")
ax6.legend(loc="upper right")

fig.suptitle('Amount of food type bought by each household in Scotland by year', fontsize=16)
plt.tight_layout()
ax5.set_xticklabels(ax5.get_xticklabels(), rotation=90, ha='right')
plt.xticks(rotation = 90)
plt.subplots_adjust(wspace=0.1, hspace=0.1)

plt.show()
```

```python
import seaborn as sns

yrs = [201516,201617,201718,201819, 201920,202021]
new_labels = ['2015/16','2016/17','2017/18','2018/19', '2019/20','2020/21']
df_ext = df[['Food Group'] + yrs]
df_ext = df_ext.groupby('Food Group').sum().reset_index()

df_melt = df_ext.melt(id_vars = ['Food Group'], var_name = 'Year', value_name = 'Quantity in grams')

plt.figure(figsize=(18,12))
sns.set_theme(style="whitegrid")

g = sns.catplot(data=df_melt, kind="bar", x = 'Food Group', y = 'Quantity in grams', hue="Year", palette="dark", alpha=.6, height=6)
# plt.legend(labels=labels)
for t, l in zip(g._legend.texts, new_labels):
    t.set_text(l)
    
plt.xticks(rotation = 'vertical', fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Food Group', fontsize = 15)
plt.ylabel('Quantity in grams', fontsize = 15)
plt.title('Annual quantity of food groups bought in \n Scottish households', fontsize = 16)
plt.show()
```
