```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Scotland_Expenditure.xlsx', sheet_name = 'Scotland', header = 7)

df = df[df['Description'] == 'Food and drink']
df = df.drop([6,11], axis=0)
df.drop(df.columns[2:5], axis=1, inplace=True)
df = df.drop(['RSE indicator(a)', '% change since 201819', 'sig(b)', 'trend since 20819(c)'], axis = 1).reset_index(drop=True)
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
plt.figure(figsize=(10, 10))

plt.plot(df.columns[2:].astype(str), df.iloc[0, 2:], marker = 'o', linestyle = '-')

plt.ylabel('Expenditure (pence/week)')
plt.xlabel('Years')
plt.title(f'Expenditure on Food and Drinks (Scotland)', fontsize = 15)
plt.grid(True)
plt.xticks(rotation = 90)
plt.show()
```
